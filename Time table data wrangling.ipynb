```python
# get yours from http://academic.iiti.ac.in/index1.html
enrolled = ["CS 303","CS 305","CS 307","CS 309","CS 353","CS 355","CS 357","CS 359"]
```

```python
relevant = pd.DataFrame(columns=tt.columns)
for col in tt.columns:
    for course in enrolled:
            relevant[col] = relevant[col] | tt[col].str.contains(course)
```

```python
tt = tt[relevant.any(axis=1)].copy(deep=True).reset_index(drop=True);
```

```python
del tt["12:00 -12:55"] # Removing lunch
```

```python
tt
```

```python
def set_col(df, i, combine):
    a = None
    for t in combine:
        if a is None:
            a = df.iloc[t].fillna("")
        else:
            a = a + df.iloc[t].fillna("")
    df.iloc[i] = a
```

```python
set_col(tt, 0, [0,1])
set_col(tt, 2, [2,3])
set_col(tt, 4, [4,5])
set_col(tt, 6, [6,7])
```

```python
tt = tt.drop([tt.index[i] for i in [1,3,5,7]]).reset_index(drop=True);
```
